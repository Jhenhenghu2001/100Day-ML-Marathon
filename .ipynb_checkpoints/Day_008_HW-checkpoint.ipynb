```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# 設定 data_path
dir_data = 'C:/Users/TL_Yu/Documents/GitHub/100Day-ML-Marathon/data'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app)
app_train.head()
```

## 作業
1. 請將 app_train 中的 CNT_CHILDREN 依照下列規則分為四組，並將其結果在原本的 dataframe 命名為 CNT_CHILDREN_GROUP
    * 0 個小孩
    * 有 1 - 2 個小孩
    * 有 3 - 5 個小孩
    * 有超過 5 個小孩

2. 請根據 CNT_CHILDREN_GROUP 以及 TARGET，列出各組的平均 AMT_INCOME_TOTAL，並繪製 baxplot
3. 請根據 CNT_CHILDREN_GROUP 以及 TARGET，對 AMT_INCOME_TOTAL 計算 [Z 轉換](https://en.wikipedia.org/wiki/Standard_score) 後的分數

```python
app_train["CNT_CHILDREN_GROUP"] = pd.cut(app_train.CNT_CHILDREN,[-1,0,2,5,max(app_train.CNT_CHILDREN)],right=True,include_lowest=True)
#app_train.loc[app_train.CNT_CHILDREN==0,'CNT_CHILDREN_GROUP'] = "0"
#app_train.loc[app_train.CNT_CHILDREN.between(1,2),'CNT_CHILDREN_GROUP'] = "1-2"
#app_train.loc[app_train.CNT_CHILDREN.between(3,5),'CNT_CHILDREN_GROUP'] = "3-5"
#app_train.loc[app_train.CNT_CHILDREN>5,'CNT_CHILDREN_GROUP'] = "5+"
pd.crosstab(index=app_train.CNT_CHILDREN,
            columns=app_train.CNT_CHILDREN_GROUP)
# app_train.CNT_CHILDREN_GROUP.value_counts()
```

```python
#2-1
grouped_df = app_train.groupby(['CNT_CHILDREN_GROUP','TARGET'])['AMT_INCOME_TOTAL']
grouped_df.mean()
```

```python
#2-2
app_train.boxplot(column='AMT_INCOME_TOTAL',#['AMT_INCOME_TOTAL','AMT_CREDIT']
                  by=['CNT_CHILDREN_GROUP','TARGET'],
                  showfliers = False,
                  figsize=(12,12))
plt.suptitle('This is Boxplot')
plt.show()
```

# 3. 
請根據 CNT_CHILDREN_GROUP 以及 TARGET，對 AMT_INCOME_TOTAL 計算 [Z 轉換](https://en.wikipedia.org/wiki/Standard_score)後的分數

```python
app_train["AMT_INCOME_TOTAL_Z_BY_CHILDREN_GRP-TARGET"]=grouped_df.apply(lambda x: (x-x.mean())/x.std())
app_train[['AMT_INCOME_TOTAL','AMT_INCOME_TOTAL_Z_BY_CHILDREN_GRP-TARGET']].head()
```
